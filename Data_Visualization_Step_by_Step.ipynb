This notebook explains data visualization from raw data to insights using Python.

## Step 1: Import Libraries

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 2: Create Sample Dataset

```python

data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [120, 150, 180, 170, 210],
    'Profit': [30, 40, 60, 55, 80]
}

df = pd.DataFrame(data)
df
```

## Step 3: Line Chart (Trend Analysis)

```python
plt.figure()
plt.scatter(df['Sales'], df['Profit'])
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.title("Sales vs Profit Relationship")
plt.show()
```

```python

plt.plot(df['Month'], df['Sales'])
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

## Step 4: Bar Chart (Comparison)

```python

plt.bar(df['Month'], df['Profit'])
plt.title('Monthly Profit')
plt.xlabel('Month')
plt.ylabel('Profit')
plt.show()
```

## Step 5: Seaborn Visualization

```python

sns.set()
sns.lineplot(x='Month', y='Sales', data=df)
plt.title('Sales Trend using Seaborn')
plt.show()
```

## Step 6: Insights

- Sales are increasing overall
- Highest sales in May
- Profit follows sales trend
